```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('data.csv')
```

## Cleaning data
To deal with empty cells i am going to use the `dropna` function with the inplace parameter set to true to remove rows with empty cells in the current data frame i'm working on.  


```python
df.dropna(inplace= True)
```

## Removing duplicate data
To remove duplicate data i am going to use the `drop_duplicates` function with the inplace parameter set to true to remove duplicate rows in the current data frame i'm working on.  

```python
df.drop_duplicates(inplace=True)
```

```python
quality_counts = df['quality'].value_counts()
labels = quality_counts.index.tolist()
sizes = quality_counts.values.tolist()

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Quality Ratings')
plt.show()
```
